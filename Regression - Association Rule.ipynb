```python
from sklearn.linear_model import LinearRegression
df = pd.read_csv('umbrellas_sold_1.csv')
att = df[['rainfall_mm']]
label = df['umbrellas_sold']

model = LinearRegression()
model.fit(att,label)

ans = model.predict(att)
plt.figure(figsize=[4,3])
plt.scatter(df['rainfall_mm'],df['umbrellas_sold'])
plt.plot(att,ans,'r--')

model.score(att,label)
#model.predict([[200]]) อยากรู้ว่าค่าน้ำฝนมีอัตตราเท่าไหร่
```

```python
from sklearn.linear_model import LinearRegression
df = pd.read_csv('electricity-consumptions.csv')
att = df[['home_size']]
label = df['kilowatt_hours_per_month']

model = LinearRegression()
model.fit(att,label)

ans = model.predict(att)
plt.figure(figsize=[4,3])
plt.scatter(df['home_size'],df['kilowatt_hours_per_month'])
plt.plot(att,ans,'r--')

model.score(att,label)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv('electricity-consumptions.csv')
att = df[['home_size']]
label = df['kilowatt_hours_per_month']

pf = PolynomialFeatures(degree =2) #degree ทำให้เส้นโค้งหาข้อมูลมากขึ้น
pf.fit(att)
att_new = pf.transform(att)

model = LinearRegression()
model.fit(att_new,label)

ans = model.predict(att_new)
plt.figure(figsize=[4,3])
plt.scatter(df['home_size'],df['kilowatt_hours_per_month'])
plt.plot(att['home_size'],ans,'r--') 

model.score(att_new,label) 
```

```python
basketData = pd.read_csv('MarketBasketOptimization.csv',header = None)
```

```python
basketData.head()
```

```python
dataset = []
for i in range(0,basketData.shape[0]) :
    temp = []
    for j in range(0,basketData.shape[1]) :
        if basketData.loc[i,j] is not np.nan :
            temp.append(basketData.loc[i,j])
    dataset.append(temp)
```

```python
basketData.shape[1] #จำนวนคอลัมน์
```

```python
dataset #รายการซื้อ
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(dataset).transform(dataset)
```

```python
te_ary
```

```python
df = pd.DataFrame(te_ary,columns = te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets = apriori(df ,use_colnames= True ,min_support = 0.01)
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets,metric='confidence',min_threshold=0)
```
