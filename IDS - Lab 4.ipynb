แบบฝึกหัดนี้เป็นแบบฝึกหัดทดสอบทักษะการใช้งาน library pandas  ด้วย [SF Salaries Dataset](https://www.kaggle.com/kaggle/sf-salaries) จากเว็ปไซต์ Kaggle ให้ทำตามคำสั่ง ต่อไปนี้

---

**Import pandas as pd.**

```python
import pandas as pd 
```

**ให้นำเข้าข้อมูลจากไฟล์ Salaries.csv มาในรูปของ dataframe โดยตังชื่อตัวแปรว่า sal**

```python
sal = pd.read_csv('Salaries.csv')
```

**Check the head of the DataFrame.**

```python
sal.head()
```

**ใช้คำสั่ง .info() method to ในการดูภาพรวมของข้อมูลทั้งหมด**

```python
sal.info()
```

**ลบคอลัมน์ Notes และ Status ออก**

```python
sal.drop('Notes',axis=1,inplace = True)
sal.drop('Status',axis=1,inplace = True)
```

```python
sal
```

**หาค่าเฉลี่ยนของ Benefits ใน sal**

```python
sal['Benefits'].mean()
```

**ใน sal แทน Benefits ที่เป็น null ด้วย 0**

```python
sal['Benefits'].fillna(value = 0,inplace = True)
```

```python
sal.head()
```

**หาค่าเฉลี่ยนของ Benefits ใน sal อีกครั้ง**

```python
sal['Benefits'].mean()
```

**จงเพิ่มคอลัมน์ Year(TH) ใน sal ให้เป็นเลขปี พศ**

```python
sal['Year(TH)'] = sal['Year'] + [543]
```

```python
sal
```

**จงเพิ่มคอลัมน์ Level มีค่าเป็น L เมื่อ TotalPayBenefits น้อยกว่า 1 แสน และเป็น H เมื่อมากกว่าเท่ากับ 1 แสน**

```python
def fn(x):
    if x < 100000 :
        return 'L'
    else :
        return 'H'
```

```python
sal['Level'] = sal['TotalPayBenefits'].apply(fn)
```

```python
sal
```

**เซ็ต Id ให้เป็น index**

```python
sal.rename( columns = {'Id':'index'} , inplace = True)
```

```python
sal
```

**เปลี่ยนชื่อคอลัมน์ Year เป็น Year(Eng)**

```python
sal.rename( columns = {'Year' : 'Year(Eng)'},inplace = True)
```

```python
sal
```

**เพิ่มคนชื่อ David Copperfield ทำงานเป็น Magician คอลัม์อื่นๆเป็น null**

```python
sal.loc[len(sal)+1] = pd.Series(['David Coperfield','Magician'],
                                index = ['EmployeeName','JobTitle'])
```

```python
sal
```

**สร้าง Dataframe ที่ EmployeeName มีนาย A , B และ C ซึ่งมี BasePay เป็น 10000 แล้วนำไปรวมกับ sal**

```python
sal2 = pd.DataFrame({ 'EmployeeName' : ['A','B','C'],
                     'BasePay' : [100000,100000,100000]
                    },index=[0,1,2])
sal = pd.concat([sal,sal2])
```

```python
sal
```

**สร้างตาราง salB ซึ่งเก็บเฉพาะของคนที่ไม่มี BasePay**

```python
salB = sal[sal['BasePay'].isnull()]
```

```python
salB.head()
```

```python
salB.info()
```
