```python
dic = {"day" :['Sun','Mon','Tue','Wed','Thu','Fir','Sat'],
      "temp1" :[24,25,29,35,23,20,23],
      "temp2" :[10,14,9,12,11,15,17]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp1']
z = df['temp2']
```

```python
plt.plot(df['day'],df['temp1'],'s-.r')
plt.plot(df['day'],df['temp2'],'o:g')
```

```python
plt.plot(df['day'],df['temp1'],'s-.r',lable = 'Temp1')
plt.plot(df['day'],df['temp2'],'o:b',lable = 'Temp2')
plt.legend(loc = 'left')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0,0].plot(df['day'],df['temp1'])
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0,0].plot(df['day'],df['temp1'],'s--r')
axes[0,1].plot(df['temp1'],df['day'],'p--y')
axes[1,0].plot(df['temp1'],df['day'],'s--k')
axes[1,1].plot(df['temp1'],df['day'],'s--g')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0,0].plot(df['day'],df['temp1'],'s--r')
axes[0,1].plot(df['temp1'],df['day'],'p--y')
axes[1,0].plot(df['temp1'],df['day'],'*-k')
axes[1,1].plot(df['day'],df['temp1'],'s--g')
axes[0,0].set_title('week1')
axes[0,1].set_title('week2')
axes[1,0].set_title('week3')
fig.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(df['day'],df['temp1'],'s--r')

axes2 = fig.add_axes([0.5,1,0.5,0.5])
axes2.plot(df['day'],df['temp1'],'s--r')
```
