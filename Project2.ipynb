```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Country-data.csv')
```

```python
df.info()
```

```python
sns.jointplot(data = df, x = 'health' , y = 'child_mort')
```

```python
sns.histplot(df['health'],bins = 20,color = 'r')
```

```python
sns.heatmap(df[['child_mort','exports','health','imports','income','inflation','life_expec','total_fer','gdpp']]
            .corr(),annot = True,cmap = 'Spectral')
plt.title('Country')
```

```python
sns.scatterplot(data = df, x = 'imports', y = 'exports', color = 'b',ec ='k',lw=1)
#imports = การนำเข้าสินค้าและค่าบริการต่อหัว
#exports = การส่งออกสินค้าและค่าบริการต่อหัว
#พอร์ตเพื่อดูการนำเข้าสินค้าและการส่งออกของสินค้า
```

```python
mean_imports = np.mean(df['imports'])
mean_exports = np.mean(df['exports'])

circle = plt.Circle((mean_imports, mean_exports), max(mean_imports, mean_exports), color='b', fill=False, linestyle='--')

plt.scatter(df['imports'], df['exports'], color='b', edgecolor='k', linewidth=1)
plt.xlabel('Imports')
plt.ylabel('Exports')
plt.title('Imports vs. Exports')
plt.gca().add_artist(circle)
plt.show()
```

```python
sns.scatterplot(data = df, x = 'child_mort', y = 'total_fer', color = 'r',ec ='k',lw=1)
#child_mort = อัตตราจำนวนเสียชีวิตของเด็ก
#total_fer = อัตตราการเกิดของเด็ก
#เป็นการพอร์ตเพื่อดูอัตตราการเสียชีวิตและอัตตราการเกินขึ้นมาใหม่
```

```python
plt.scatter(data = df, x = 'income', y = 'gdpp',c= 'pink',ec ='k',lw=1 )
#income = รายได้สุทธิต่อบุคคล
#gdpp = เป็นตัวชี้วัดผลการดำเนินงานของเศรษฐกิจโดยรวม
#เป็นการพล็อตเพื่อดูความสัมพันธ์ระหว่างรายได้กับ gdp โดยมีขอบเขตรายได้ตั้งแต่ 0 ถึง 100,000 
#เพื่อที่ได้รู้ว่ารายได้เท่านี้จะมี gdp เท่าไหร่หรือรายได้เท่านี้จะมีรายได้ที่สูงขึ้น ค่า gdp ก็มีแนวโน้มสูงขึ้นตามไปด้วย
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = df['child_mort'].values.reshape(-1, 1)
y = df['total_fer'].values.reshape(-1, 1)
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='child_mort', y='total_fer', data=df, hue='income', palette='Spectral')
plt.plot(x, y_pred, color='k')
plt.title('Child Mortality vs. Total fertility')
plt.xlabel('Child Mortality')
```
