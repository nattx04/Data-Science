```python
x = ['Mon','Tue','Wed','Thu','Fri']
y = [50,70,35,90,40]
```

```python
plt.bar(x,y,color = 'b',alpha = 0.5,hatch = '\/-|')
```

```python
plt.figure(figsize = [5,3])
plt.bar(x,y,color = 'r',alpha = 0.5,hatch = 'xx'
       ,lw = 2,ec = 'k',ls = ':')
```

```python
plt.figure(figsize = [4,2])
plt.bar(x,y,color = 'r',alpha = 0.5,hatch = 'xx'
       ,lw = 2,ec = 'k',ls = ':')
plt.plot(x,y)
plt.title('week')
```

```python
plt.figure(figsize = [4,2])
plt.barh(x,y,color = 'g',alpha = 0.5,hatch = 'xx'
       ,lw = 2,ec = 'k',ls = ':')
plt.title('week')
```

```python
x = np.arange(1,6)
y1 = np.random.randint(1,50,5)
y2 = np.random.randint(1,50,5)
y3 = np.random.randint(1,50,5)
```

```python
plt.figure(figsize = [4,2])
bw = 0.2
plt.bar(x,y1,width = bw)
plt.bar(x + bw,y2,width = bw)
plt.bar(x + bw*2,y3,width = bw)
```

```python
plt.figure(figsize = [4,2])
bw = 0.2
plt.bar(x,y1,width = bw)
plt.bar(x + bw,y2,width = bw)
plt.bar(x + bw*2,y3,width = bw)
plt.xticks(x + bw,[2000,2001,2002,2003,2004])
plt.show()
```

```python
plt.figure(figsize = [4,2])
bw = 0.2
plt.bar(x,y1,width = bw,label ='A')
plt.bar(x + bw,y2,width = bw,label ='B')
plt.bar(x + bw*2,y3,width = bw,label ='C')
plt.xticks(x + bw,[2000,2001,2002,2003,2004])
plt.legend(loc = 'best')
plt.show()
```

```python
dic = {"Math":[10,40,78,55,60],
      "Eng":[99,23,56,97,65],
      "Thai":[78,50,60,34,25]
      }
df = pd.DataFrame(dic)
data = [df['Math'],df['Eng'],df['Thai']]
plt.boxplot(data,vert = True,widths = 0.3,labels = ["Math","Eng","Thai"]
            ,patch_artist = True)
plt.show
```
